```python
# Settings and packages
%matplotlib inline 
import matplotlib.pyplot as plt # package for plotting
import numpy as np # package for numerical analyis
```

```python
# Import qutip package
from qutip import *
from qutip.qip.operations import sqrtnot, y_gate
from qutip.qip.circuit import sqrtnot, y_gate
```

```python
sqX = sqrtnot()
sqX.conj()
```

```python
Y = y_gate()
Y
```

```python
U = sqX*Y*(sqX.conj())
U
```

```python
N=2 # Dimension of Hilbert space

up = basis(N,0)  # sigma_z eigenstate |up>=(1,0)
dn = basis(N,1)  # sigma_z eigenstate |dn>=(0,1)

psi1 = up
psi2 = (up+dn)/np.sqrt(2)
psi3 = (up+1j*dn)/np.sqrt(2)
```

```python
phi1 = U*psi1
phi1
```

```python
phi2 = U*psi2
phi2
```

```python
phi3 = U*psi3
phi3
```

```python
sphere = Bloch()
sphere.add_states(psi1)
sphere.show()
```

```python
g=2
muB=0.0028
t = 2.8
B = 100

sphere = Bloch()
sphere.add_states(psi1)

i = 3

while (i>0):
    if (i == 3):
        H = 0.5*g*muB*B*sigmax()
        tlist = np.linspace(0.0,t,100)
        psi_init = psi1
        result1 = mesolve(H,psi1,tlist,[],[sigmax(),sigmay(),sigmaz()])
        result_to_plot = result1
        sphere.add_points(result_to_plot.expect,meth='l')
        sphere.add_states(sqX*psi1)
        sphere.show()
        
    elif (i == 2):
        H = 0.5*g*muB*B*sigmay()
        tlist = np.linspace(0.0,2*t,100)
        psi_init = sqX*psi1
        result2 = mesolve(H,sqX*psi1,tlist,[],[sigmax(),sigmay(),sigmaz()])
        result_to_plot = result2
        sphere.add_points(result_to_plot.expect,meth='l')
        sphere.add_states(Y*sqX*psi1)
        sphere.show()
        
    elif (i == 1):
        H = -0.5*g*muB*B*sigmax()
        tlist = np.linspace(0.0,t,100)
        psi_init = Y*sqX*psi1
        result3 = mesolve(H,Y*sqX*psi1,tlist,[],[sigmax(),sigmay(),sigmaz()])
        result_to_plot = result3
        sphere.add_points(result_to_plot.expect,meth='l')
        sphere.add_states((sqX.conj())*Y*sqX*psi1)
        sphere.show()
        
    i -= 1
```
